```python
sales_df = pd.read_csv('sales_scaled.csv')
sales_df
```

```python
sales_df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(sales_df.drop('flag',axis=1),sales_df['flag'], 
                                                  test_size=0.2, random_state=123)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp_clf = MLPClassifier(hidden_layer_sizes=(10,10), max_iter=300)
mlp_clf.fit(X_train , y_train)
mlp_pred = mlp_clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score , recall_score, accuracy_score

def get_clf_eval(y_test,pred):
    confusion = confusion_matrix(y_test,pred)
    accuracy = accuracy_score(y_test,pred)
    recall = recall_score(y_test,pred)
    precision = precision_score(y_test,pred)
    print('confusion matrix')
    print(confusion)
    print("정확도: {0:.4f}. 재현율: {1:.4f}, 정밀도: {2:.4f}".format(accuracy, recall, precision))
```

```python
get_clf_eval(y_test,mlp_pred)
```

```python
from keras import models
from keras.layers import Input, Dense
```

```python
# 모델 정의
model=models.Sequential()
model.add(Input(shape=(X_train.shape[1],)))
model.add(Dense(10, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
print(model.summary())
```

```python
# 학습 정의
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 학습하기
history=model.fit(X_train,y_train,
                 epochs=200,batch_size=128,
                 validation_split=0.3)
```

```python
# 학습 결과 확인하기
```

```python
history_dict=history.history
print(history_dict.keys())
```

```python
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']
epochs=range(1,len(acc)+1)
```

```python
import matplotlib.pyplot as plt 

# Loss 함수 그리기
plt.plot(epochs,loss,'bo',label='Training loss')
plt.plot(epochs,val_loss,'b',label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.legend()
plt.show()
```

```python
# 정확도 그리기
plt.plot(epochs,acc,'bo',label='Training accuracy')
plt.plot(epochs,val_acc,'b',label='Validation accuracy')
plt.title('Training and validation Accuracy')
plt.xlabel('Epochs')
plt.legend()
plt.show()
```
