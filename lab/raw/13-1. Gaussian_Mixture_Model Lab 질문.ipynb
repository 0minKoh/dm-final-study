군집화 전에 medv 컬럼을 제외한 컬럼들에 대한 Min-Max Scaling을 수행하시오.

```python
# from sklearn.datasets import load_boston
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
boston_df = pd.read_csv('../../data/boston_housing.csv')
boston_df
```

```python
boston_scaled = MinMaxScaler().fit_transform(boston_df.drop('medv', axis=1))
boston_scaled_df=pd.DataFrame(data= boston_scaled,columns=boston_df.columns[:-1])
```

```python
boston_scaled_df
```

```python
### Function to visualize clustering results  
import numpy as np
def visualize_cluster_plot(clusterobj, dataframe, label_name, iscenter=True):
    if iscenter :
        centers = clusterobj.cluster_centers_
        
    unique_labels = np.unique(dataframe[label_name].values)
    markers=['o', 's', '^', 'x', '*','P','D','H','v','>','<']
    isNoise=False

    for label in unique_labels:
        label_cluster = dataframe[dataframe[label_name]==label]
        if label == -1:
            cluster_legend = 'Noise'
            isNoise=True
        else :
            cluster_legend = 'Cluster '+str(label)
        
        plt.scatter(x=label_cluster['ftr1'], y=label_cluster['ftr2'], s=70,\
                    edgecolor='k', marker=markers[label], label=cluster_legend)
        
        if iscenter:
            center_x_y = centers[label]
            plt.scatter(x=center_x_y[0], y=center_x_y[1], s=250, color='white',
                        alpha=0.9, edgecolor='k', marker=markers[0])  #label
            plt.scatter(x=center_x_y[0], y=center_x_y[1], s=70, color='k',\
                        edgecolor='k', marker='$%d$' % label)
    if isNoise:
        legend_loc='upper center'
    else: legend_loc='upper right'
    
    plt.legend(loc=legend_loc)
    plt.show()
```

```python
# boston_housing.csv 파일을 boston_df 데이터 프레임으로 만들어서 medv 컬럼을 제외한 컬럼들을 이용해서 GMM 알고리즘을 적용할 결과를 PCA를 적용한 후 visualize_cluster_plot() 함수를 이용하여 가시화하시오(단 군집의 수는 7개로 하시오).
# 군집화 전에 medv 컬럼을 제외한 컬럼들에 대한 Min-Max Scaling을 수행하시오.
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture

# GMM 알고리즘을 적용할 결과를 PCA를 적용한 후 visualize_cluster_plot() 함수를 이용하여 가시화하시오(단 군집의 수는 7개로 하시오)
# 3개의 n_components기반 GMM을 X_aniso 데이터 셋에 적용
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(boston_scaled_df) # irisDF.iloc[:, :4]와 동일
boston_df['ftr1'] = pca_transformed[:,0]
boston_df['ftr2'] = pca_transformed[:,1]

gmm = GaussianMixture(n_components=3, random_state=0)
gmm_label = gmm.fit(boston_scaled_df).predict(boston_scaled_df)
boston_df['gmm_label'] = gmm_label

# GaussianMixture는 cluster_centers_ 속성이 없으므로 iscenter를 False로 설정. 
visualize_cluster_plot(gmm, boston_df, 'gmm_label',iscenter=False)
```
